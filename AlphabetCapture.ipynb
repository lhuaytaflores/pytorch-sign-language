```python
import cv2
import os
import imutils
```

```python
videonames = ['A_L', 'A_R', 'B_L', 'B_R', 'C_L', 'C_R', 'D_L', 'D_R', 'E_L', 'E_R', 'F_L', 'F_R', 'G_L', 'G_R', 'H_L', 'H_R', 'I_L', 'I_R', 'J_L', 'J_R', 'K_L', 'K_R', 'L_L', 'L_R', 'M_L', 'M_R', 'N_L', 'N_R', 'O_L', 'O_R', 'P_L', 'P_R', 'Q_L', 'Q_R', 'R_L', 'R_R', 'S_L', 'S_R', 'T_L', 'T_R', 'U_L', 'U_R', 'V_L', 'V_R', 'W_L', 'W_R', 'X_L', 'X_R', 'Y_L', 'Y_R', 'Z_L', 'Z_R']
alphabetfolder = ['A_L', 'A_R', 'B_L', 'B_R', 'C_L', 'C_R', 'D_L', 'D_R', 'E_L', 'E_R', 'F_L', 'F_R', 'G_L', 'G_R', 'H_L', 'H_R', 'I_L', 'I_R', 'J_L', 'J_R', 'K_L', 'K_R', 'L_L', 'L_R', 'M_L', 'M_R', 'N_L', 'N_R', 'O_L', 'O_R', 'P_L', 'P_R', 'Q_L', 'Q_R', 'R_L', 'R_R', 'S_L', 'S_R', 'T_L', 'T_R', 'U_L', 'U_R', 'V_L', 'V_R', 'W_L', 'W_R', 'X_L', 'X_R', 'Y_L', 'Y_R', 'Z_L', 'Z_R']

videofolder = 'video_data'

totalframes = 160
```

```python
_pathvideo = '/' + videofolder + '/'   
_pathtrain = '/dataset/train/'
_pathval = '/dataset/val/'

i = 0

for videoname in videonames:
    pathvideo = _pathvideo + videoname + '.mp4'
    pathtrain = _pathtrain + alphabetfolder[i] + '/'
    pathval = _pathval + alphabetfolder[i] + '/'
    if not os.path.exists(pathtrain):
        print('Created trainning folder: ', pathtrain)
        os.makedirs(pathtrain)
    if not os.path.exists(pathval):
        print('Created validation folder: ', pathval)
        os.makedirs(pathval)
    i += 1
    capture = cv2.VideoCapture(pathvideo)
    countframes = 0
    while True:
        ret, frame = capture.read()
        if ret == True:
            countframes += 1
        else:
            capture.release()
            break
    print('Number of video frames ' + videoname + '.mp4: ' + str(countframes))
    
    capture = cv2.VideoCapture(pathvideo)
    flag = (int)(countframes / totalframes)
    count = 0
    numberframes = 0
    while True:
        ret, frame = capture.read()
        if (ret == True):
            frame = imutils.resize(frame, width = 250)
            cv2.imshow('Video', frame)
            if (cv2.waitKey(1) == ord('s')):
                break
            numberframes += 1
            if numberframes % flag == 0 and count < totalframes:
                width = frame.shape[1]
                heigth = frame.shape[0]
                matrix = cv2.getRotationMatrix2D((width / 2, heigth / 2), 360, 1)
                output = cv2.warpAffine(frame, matrix, (width, heigth))
                if count + 1 <= totalframes * 0.8:
                    cv2.imwrite(pathtrain + videoname + str(count) + '.jpg', output)
                else:
                    cv2.imwrite(pathval + videoname + str(count) + '.jpg', output)
                count += 1
        else:
            print('Number of video frames created ' + videoname + '.mp4: ' + str(count))
            break

capture.release()
cv2.destroyAllWindows()
```
